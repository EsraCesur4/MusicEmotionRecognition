```python
import pandas as pd
import numpy as np
import torch
```

# Data

```python
df = pd.read_csv("https://raw.githubusercontent.com/Ezeysay16/veridosyalari/main/train_set.csv")
df.head()
```

```python
columns = ["id","name", "artist","realease_year", "key", "mode", "time_signature", "duration_ms"]
for col in columns:

  del df[col]
df.head()
```

```python
labelled = df[df["emotion_tag"].notnull()]
labelled.head()
```

```python
unlabelled = df[df["emotion_tag"].isnull()]
unlabelled.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Creating a instance of label Encoder.
le = LabelEncoder()

# Using .fit_transform function to fit label
# encoder and return encoded label
label = le.fit_transform(labelled['emotion_tag'])

# printing label
label

labelled.drop("emotion_tag", axis=1, inplace=True)

# Appending the array to our dataFrame
# with column name 'Purchased'
labelled["emotion_tag"] = label

# printing Dataframe
labelled
```

```python
test = pd.read_csv("https://raw.githubusercontent.com/Ezeysay16/veridosyalari/main/test_set_no_labels.csv")
test.head()
```

```python
columns = ["name", "artist","realease_year", "key", "mode", "time_signature", "duration_ms"]
for col in columns:
  del test[col]
test.head()
```

# Model

```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
```

```python
# Load your labeled and unlabeled data as pandas DataFrames
labeled_data = labelled
unlabeled_data = unlabelled

# Perform pseudo-labelling on the unlabeled data
pseudo_labels = []

# Define the Naive Bayes classifier
naive_bayes = GaussianNB()

x_labeled = labeled_data[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
       'instrumentalness', 'valence']]
y_labeled = labeled_data['emotion_tag']

x_unlabeled = unlabeled_data[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
       'instrumentalness', 'valence']]

# Train the Naive Bayes classifier on the labeled data
naive_bayes.fit(x_labeled, y_labeled)

# Generate pseudo-labels for the unlabeled data
pseudo_labels = naive_bayes.predict(x_unlabeled)

# Combine the labeled and pseudo-labeled data

X_combined = np.concatenate((x_labeled, x_unlabeled), axis=0)
y_combined = np.concatenate((y_labeled, torch.tensor(pseudo_labels)), axis=0)
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X_combined, y_combined, test_size=0.2, random_state=42)

X_train = X_combined
y_train = y_combined
X_test = test[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
       'instrumentalness', 'valence']]
y_test = test["emotion_tag"]

# Normalize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convert data to PyTorch tensors
X_train = torch.tensor(X_train, dtype=torch.float)
y_train = torch.tensor(y_train, dtype=torch.long)
X_test = torch.tensor(X_test, dtype=torch.float)


# Define the MLP model architecture
class MLP(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        return out

# Set the hyperparameters
input_size = X_train.shape[1]
hidden_size = 64
num_classes = 4
learning_rate = 0.001
num_epochs = 10
batch_size = 32

# Create the MLP model
model = MLP(input_size, hidden_size, num_classes)

# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    model.train()
    total_loss = 0.0
    total_samples = 0

    # Mini-batch training
    for i in range(0, len(X_train), batch_size):
        # Get the mini-batch data
        inputs = X_train[i:i+batch_size]
        labels = y_train[i:i+batch_size]

        # Forward pass
        outputs = model(inputs)
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Update loss and sample count
        total_loss += loss.item() * inputs.size(0)
        total_samples += inputs.size(0)

    # Calculate average loss
    avg_loss = total_loss / total_samples

    # Print the average loss for this epoch
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}")

# Evaluation
model.eval()
with torch.no_grad():
    # Predict on test set
    outputs = model(X_test)
    _, predicted = torch.max(outputs.data, 1)

    # Calculate accuracy

    #print(f"Test Accuracy: {accuracy:.4f}")
```
